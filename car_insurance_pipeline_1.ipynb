```python
remove_z_cols = ['MSTATUS', 'GENDER', 'EDUCATION', 'OCCUPATION', 'CAR_TYPE', 'URBANICITY']
remove_z_cols
```

```python
def remove_z(x):

    if type(x) == float:
        return x

    elif 'z_' in x:
        return x.replace('z_', '')
    else:
        return x
    
for col in remove_z_cols:

    df[col] = df[col].apply(remove_z)

df[['MSTATUS', 'GENDER', 'EDUCATION', 'OCCUPATION', 'CAR_TYPE', 'URBANICITY']]
```

```python
def clean_num(x):

    if type(x) == float:
        return x
    
    elif ('$' in x) or (',' in x ):
        return x.replace('$', '').replace(',', '')

    else:
        return x

for col in num_cols:

    df[col] = df[col].apply(clean_num).astype('float')

df[['INCOME', 'HOME_VAL', 'BLUEBOOK', 'OLDCLAIM', 'CLM_AMT']]
```

# Data Preprocessing Steps for Machine Learning :

* 1- Split Data into Input Features and Target Column

* 2- Split Data into Train & Test

* 3- Numerical Cols : Impute Missing --> Scaling

* 4- Categorical Cols : Impute Missing --> Encoding

* 5- Handle Imbalance

### 1- Split Data into input Features and Target Feature

```python
x = df.drop('CLAIM_FLAG', axis= 1)
y = df['CLAIM_FLAG']
```

```python
df['CLAIM_FLAG'].value_counts()
```

## Numerical pipeline

```python
num_cols = x.select_dtypes(include= 'number').columns
num_cols
```

```python
from sklearn.pipeline import Pipeline
from sklearn.impute import KNNImputer
from sklearn.preprocessing import RobustScaler

knn_imputer = KNNImputer()
rc = RobustScaler()

num_pipeline = Pipeline([('Knn', knn_imputer), ('Scaling', rc)])
num_pipeline
```

## Categorical Pipelines

### Occupation Column Pipeline (Imputation --> Encoding)

```python
from sklearn.impute import SimpleImputer
from category_encoders import BinaryEncoder

be = BinaryEncoder()
imputer = SimpleImputer(strategy= 'constant', fill_value= 'Other')

occupation_pipeline = Pipeline([('Cat Impute Pipeline', imputer), ('BE', be)])
occupation_pipeline
```

### OHE Pipeline

```python
from sklearn.preprocessing import OneHotEncoder

ohe = OneHotEncoder(drop= 'first', sparse_output= False)

ohe_pipeline = Pipeline([('OHE Pipeline', ohe)])
ohe_pipeline
```

### Ordinal Pipeline

```python
from sklearn.preprocessing import OrdinalEncoder

ord = OrdinalEncoder(categories= [['<High School', 'High School', 'Bachelors', 'Masters', 'PhD']])

ord_pipeline = Pipeline([('Ordinal Pipeline', ord)])
ord_pipeline
```

### Assign each column to the corresponding Pipeline

```python
ohe_cols = x.select_dtypes(include= ['object']).columns.drop(['EDUCATION', 'OCCUPATION'])
ohe_cols
```

```python
from sklearn.compose import ColumnTransformer

Preprocessing = ColumnTransformer(transformers= [ ('Numerical', num_pipeline, num_cols),
                                                  ('Occupation Categorical', occupation_pipeline, ['OCCUPATION']),
                                                  ('OHE', ohe_pipeline, ohe_cols),
                                                  ('Ordinal', ord_pipeline, ['EDUCATION'])] )
Preprocessing
```

```python
from imblearn.pipeline import Pipeline
from imblearn.over_sampling import SMOTE
from sklearn.model_selection import cross_validate

from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from catboost import CatBoostClassifier
from lightgbm import LGBMClassifier

models = [
    ('Logistic Regression', LogisticRegression(random_state= 42)),
    ('KNN', KNeighborsClassifier()),
    ('Naive Bayes', GaussianNB()),
    ('Decision Tree', DecisionTreeClassifier(random_state= 42)),
    ('Random Forest', RandomForestClassifier(random_state= 42)),
    ('XGBoost', XGBClassifier()),
    ('CatBoost', CatBoostClassifier()),
    ('LightGBM', LGBMClassifier())
]

for model in models:

    model_pipeline = Pipeline(steps= [ ('Preprocessing', Preprocessing),
                                       ('smote', SMOTE()),
                                       ('Model', model[1])])
    
    result = cross_validate(model_pipeline, x, y, scoring= 'f1', cv= 5, return_train_score= True, n_jobs= -1)

    print(model[0])
    print('Train Score :', round(result['train_score'].mean() * 100, 2))
    print('Test Score :', round(result['test_score'].mean() * 100, 2))
    print('-' * 50)
```

```python
from lightgbm import LGBMClassifier
from sklearn.model_selection import RandomizedSearchCV

model_pipeline = Pipeline(
    steps=[
        ('Preprocessing', Preprocessing),
        ('smote', SMOTE()),
        ('Model', LGBMClassifier(
            random_state=42,
            verbose=-1
        ))
    ]
)

params = {
    'Model__n_estimators': [100, 200, 300],
    'Model__learning_rate': [0.01, 0.03, 0.05, 0.1],
    'Model__max_depth': [3, 5, 7, 10],
    'Model__num_leaves': [15, 31, 63]
}

random_search = RandomizedSearchCV(
    estimator=model_pipeline,
    param_distributions=params,
    n_iter=13,
    scoring='f1',
    cv=5,
    random_state=42,
    n_jobs=-1,
    return_train_score=True
)

random_search.fit(x, y)

print(random_search.best_score_)
print(random_search.best_params_)
```
